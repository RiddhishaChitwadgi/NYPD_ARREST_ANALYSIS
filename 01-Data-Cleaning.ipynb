```python
import pandas as pd
import numpy as np
```

```python
#Load dataset into dataframe
df = pd.read_csv('NYPD_Arrest_Data__Year_to_Date_.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#Check null values
df.isnull().sum()
```

```python
#Convert datatypes
df.convert_dtypes()
```

```python
#Dropping columns
df = df.drop(columns=['ARREST_KEY', 'PD_DESC', 'ARREST_PRECINCT', 'JURISDICTION_CODE', 'X_COORD_CD', 'Y_COORD_CD', 'KY_CD', 'PD_CD'])
```

```python
#renaming columns
df = df.rename(columns={'ARREST_DATE': 'Arrest Date', 'LAW_CODE': 'Law Code', 'OFNS_DESC': 'Offense description', 'AGE_GROUP': 'Age Group', 'PERP_SEX': 'Perpetrator Sex', 'PERP_RACE': 'Perpetrator Race', 'LAW_CAT_CD': 'Offense Level', 'ARREST_BORO': 'Borough'})
```

```python
df
```

```python
#converting datatypes
df['Arrest Date'] = pd.to_datetime(df['Arrest Date'])
```

```python
#Replacing abreviations
df['Perpetrator Sex'] = df['Perpetrator Sex'].replace(['F', 'M'], ['Female', 'Male'])
df['Borough'] = df['Borough'].replace(['B', 'S', 'K', 'M', 'Q'], ['Bronx', 'Staten island', 'Brooklyn', 'Manhattan', 'Queens'])
df['Offense Level'] = df['Offense Level'].replace(['F', 'M', 'V'], ['Felony', 'Misdemeanor','Violation'])
```

```python
#REMOVING ROWS WITH MISSING VALUES
df = df.dropna()
```

```python
#REMOVING DUPLICATE VALUES
df = df.drop_duplicates()
```

```python
#Save the cleaned dataset
df.to_csv('NYPD_Arrest_Data__Year_to_Date_cleaned.csv', index=False)
```
